```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

url = r'C:\Users\Eduardo Fonteneli\Desktop\Biblioteca\PROGRAMAÇÃO\Cientista de Dados - EBAC\Módulo 6/b1abb33c4f.csv'
df1 = pd.read_csv(url)
df1
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 3)
sns.countplot(ax = axes[0, 0], x='APGAR1', data=df1)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=df1)
sns.countplot(ax = axes[0, 2], x='SEXO', data=df1)
sns.countplot(ax = axes[1, 0], x='GESTACAO', data=df1)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=df1)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
categorias=['0', '1 a 0'
            , '4 a 0'
            , '7 ou mais', '-']
df1['CONSULTAS']= (df1['CONSULTAS']
                   .astype('category')
                  .cat.rename_categories(categorias))
ax=sns.catplot(data=df1
              , x='CONSULTAS'
              , y='APGAR5'
              , kind='bar')
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(12, 8))
fig, axes = plt.subplots(2,2 )
sns.countplot(ax = axes[0, 0], x='SEXO', data=df1)
sns.countplot(ax = axes[0, 1], x='CONSULTAS', data=df1)
sns.countplot(ax = axes[1, 0], x='GESTACAO', data=df1)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=df1)
```
